# Exercise B-6: Deterministic chaos.
A simple linear iteration of the form
$$
x_{n+1} := ax_n, \ \ \ n = 0, 1, 2, \ . \ . \ .
$$
with a given initial value $x_0 > 0$ and growth parameter $a > 1$ describes exponential growth: The solution $x_n = a^nx_0$ tends to infinity as $n \Rightarrow \infty$.
A more useful mathematical model for a growth process includes a bound that cannot be exceeded by $x_n$. To achieve this, the iteration is modified such that the growth rate decreases as the sequence approaches the bound.
We choose 1 as the bound:
$$
x_{n+1} := a(1 âˆ’ x_n)x_n, \ \ \ \ n = 0, 1, 2, . . . , a \in (1, 4), x_0 \in (0, 1). \ \ \ \ (B.1)
$$
This iteration is nonlinear, and no simple solution formula can be written down.

1.  Show: For $x_0 \in (0, 1)$ and $a \in (1, 4)$, we have $x_n \in (0, 1)$ for all
    $n = 1, 2, 3, \ . \ . \ .$.
    Hint: For $n = 0$, we assume $x_0 \in (0, 1)$. Show that if $x_n \in (0, 1)$ for
    some $n$, then $x_{n+1} \in (0, 1)$. (The statement holds for $n = 1$, and thus
    for $n = 2$, $n = 3$, etc. This type of reasoning is called induction.)

2.  How do the values $x_n$ behave?\
Choose an initial value $x_0$ and calculate enough values of $x_n$, restarting
the iteration and increasing the value of $a$ by steps of $\frac{1}{100}$, starting
from $a = 1$. For each value $a_j = 1 +\frac{j}{100}$, you obtain a sequence of iterates $x0, x1, x2, x3, \ . \ . \ .$. Plot all the points $(a_j, x_n)$ in a two-dimensional diagram. The resulting picture illustrates the behavior of the sequence $(x_n)$ as a function of the parameter $a$. For $a \Rightarrow 4$, the behavior becomes completely chaotic (this has nothing to do with rounding errors), even though the system is deterministic as defined by (B.1). This yields an interesting plot.


### Solution for 1

#### Starting point
- We want to prove:
$$
x_{n+1} := ax_n
$$
- Additionally we know:
$$
a \in (1,4), \ x_0 \in (0,1)
$$

#### Show $x_1 \in (0,1)$
$$
x_0 \in (0,1) \Rightarrow (1-x_0) \in (0,1) \Rightarrow (1-x_0)x_0 \in (0,1)
$$
$$
(1-x_0)x_0 = x_0 - x_0^2 
$$
- This function has a maximum at $x_0 = \frac{1}{2}$ with $f(\frac{1}{2}) = \frac{1}{4} \Rightarrow (1-x_0)x_0 \in (0,\frac{1}{4}]$
- Now lets say $y = (1-x_0)x_0$
- We know $a \in (1,4)$ and $y \in (0,\frac{1}{4}]$
- Hence we see $a\cdot y \in (0,1)$ as $a \neq 4 \Rightarrow x_1 = a(1-x_0)x_0 \in (0,1)$


#### Show $x_{n+1} \in (0,1)$
- Since $x_1 \in (0,1)$ we assume $x_n \in (0,1)$
- Now we just need to prove this for $x_{n+1}$
$$
x_n \in (0,1) \Rightarrow (1-x_n) \in (0,1) \Rightarrow (1-x_n)x_n \in (0,1)
$$
$$
(1-x_n)x_n = x_n - x_n^2 
$$
- This function has a maximum at $x_n = \frac{1}{2}$ with $f(\frac{1}{2}) = \frac{1}{4} \Rightarrow (1-x_n)x_n \in (0,\frac{1}{4}]$
- Now lets say $y = (1-x_n)x_n$
- We know $a \in (1,4)$ and $y \in (0,\frac{1}{4}]$
- Hence we see $a\cdot y \in (0,1)$ as $a \neq 4 \Rightarrow x_{n+1} = a(1-x_n)x_n \in (0,1)$



### Solution for 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
x0 = 0.5          # Initial value
a_values = np.linspace(1, 4, 300)  # 300 steps for a in [1, 4]
a_values_detail = np.linspace(3, 4, 200)  # 300 steps for a in [3, 4]
iterations = 500   # Total iterations for each a
transient = 100    # Number of transient iterations to discard
# Data for plotting
a_list = []
x_list = []
```

```python
# Iterate over different values of a
for a in a_values:
    x = x0
    # Iterate the sequence
    for n in range(iterations):
        x = a * (1 - x) * x
        # Record values after transient iterations
        if n >= transient:
            a_list.append(a)
            x_list.append(x)
```

```python
# Plot the bifurcation diagram
plt.figure(figsize=(10, 6))
plt.scatter(a_list, x_list, s=0.1, color="black", alpha=0.5)
plt.title("Bifurcation Diagram")
plt.xlabel("Parameter a")
plt.ylabel("Values of $x_n$")
plt.grid(True)
plt.show()
```

```python
# Data for plotting
a_list_2 = []
x_list_2 = []

# Iterate over different values of a
for a in a_values_detail:
    x = x0
    # Iterate the sequence
    for n in range(iterations):
        x = a * (1 - x) * x
        # Record values after transient iterations
        if n >= transient:
            a_list_2.append(a)
            x_list_2.append(x)
```

```python
# Plot the bifurcation diagram again
plt.figure(figsize=(10, 6))
plt.scatter(a_list_2, x_list_2, s=0.1, color="black", alpha=0.5)
plt.title("Bifurcation Diagram")
plt.xlabel("Parameter a")
plt.ylabel("Values of $x_n$")
plt.grid(True)
plt.show()
```
